# Training and testing the Data

```python
X=df3.drop("Churn",axis=1)
y=df3["Churn"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Creating ANN Model

```python
import tensorflow as tf
```

```python
model=tf.keras.Sequential(
    #Here, 64 is no of hidden layers, input_shape is the total no. of input columns
      [tf.keras.layers.Dense(64,input_shape=(27,),activation="relu"),
      tf.keras.layers.Dense(1,activation='sigmoid')]  #output layer . 1 for binary class, 3 for multiclass
)
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_predicted=model.predict(X_test)
```

```python
y_predicted=[1 if i >0.5 else 0 for i in y_predicted]
```
